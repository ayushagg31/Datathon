```python
import os
import pandas as pd
import numpy as np
from sklearn import preprocessing,metrics 
from IPython.core.display import HTML
pd.set_option("display.max_columns",75)
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import linear_model,svm
from sklearn.metrics import average_precision_score
from sklearn.metrics import precision_recall_curve
```

```python
df_raw = pd.read_csv('/home/gozmit/Downloads/lending_club_loans_data.csv')
```

```python
df_actual=df_raw['actual']
df_predicted=df_raw['predicted']
df = df_raw
df.drop(['loan_status','addr_state','earliest_cr_line','sub_grade','fico_range_low','last_credit_pull_d','last_fico_range_high','last_fico_range_low','predicted'],axis=1,inplace=True)
```

```python
Final_data = df
Final_data.fillna(0,inplace = True)
Final_data['grade'] = Final_data['grade'].map({'A':7,'B':6,'C':5,'D':4,'E':3,'F':2,'G':1})
Final_data["home_ownership"] = Final_data["home_ownership"].map({"MORTGAGE":6,"RENT":5,"OWN":4,"OTHER":3,"NONE":2,"ANY":1})
Final_data["emp_length"] = Final_data["emp_length"].replace({'years':'','year':'',' ':'','<':'','\+':'','n/a':'0'}, regex = True)
#Final_data["emp_length"] = Final_data["emp_length"].replace({'years':'','year':'',' ':'','<':'','\+':'','n/a':'0'}, regex = True)
Final_data["emp_length"] = Final_data["emp_length"].apply(lambda x:int(x))
Final_data["purpose"] = Final_data["purpose"].map({"major_purchase":1,"car":1,"wedding":1,"credit_card":1,"home_improvement":2,"vacation":2,"debt_consolidation":2,"house":3,"moving":3,"medical":3,"other":3,"renewable_energy":3,"educational":4,"small_business":4})
print("Current shape of dataset :",Final_data.shape)
Final_data
```

```python
scl = preprocessing.StandardScaler() #instance of preprocessing
fields = Final_data.columns.values[:-1]
data_clean = pd.DataFrame(scl.fit_transform(Final_data[fields]), columns = fields)
data_clean['actual'] = Final_data['actual']
data_clean['actual'].value_counts()
```

```python
data_clean.corr()
```

```python
loanstatus_0 = data_clean[data_clean["actual"]==0]
loanstatus_1 = data_clean[data_clean["actual"]==1]
subset_of_loanstatus_0 = loanstatus_0.sample(n=6448)
subset_of_loanstatus_1 = loanstatus_1.sample(n=8000)
data_clean = pd.concat([subset_of_loanstatus_1, subset_of_loanstatus_0])
data_clean = data_clean.sample(frac=1).reset_index(drop=True)
print("Current shape of dataset :",data_clean.shape)
data_clean.head()
```
